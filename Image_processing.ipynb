## READ AND WRITE AN IMAGE

```python
# image read, image write -> opencv
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread(r'C:\Users\gopin\Desktop\Tutorial\Pycharm_files\1.jpg')
```

```python
# cv2.imshow('Image',image)
# cv2.waitKey(1)
plt.imshow(image)
plt.show()
```

```python
# image shape
image.shape # 409 -> height, 612 -> width, 3-> no. of color channels 
```

```python
# image write
cv2.imwrite('output.jpg',image)
```

```python
# image Resize
resized_image = cv2.resize(image,(300,300))
plt.imshow(resized_image)
plt.show()
```

```python
# Erode image
import numpy as np
kernal = np.ones((5,5),np.uint8) # filter
kernal
```

```python
erode_image = cv2.erode(image,kernal)
plt.imshow(erode_image)
plt.show()
```

```python
# image conversion to gray scale
gray_image = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray_image)
plt.show()
```

```python
gray_image.shape
```

```python
# Thresholding or binary conversion 0-255
sucess,threshold_image = cv2.threshold(gray_image,120,255,cv2.THRESH_BINARY)
plt.imshow(threshold_image)
plt.show()
```
